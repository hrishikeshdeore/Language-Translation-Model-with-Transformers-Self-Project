#Preprocessing the data

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
tokenizer("Hello, this is a sentence!")
```

```python
tokenizer(["Hello, this is a sentence!", "This is another sentence."])
```

```python
with tokenizer.as_target_tokenizer():
    print(tokenizer(["एक्सेर्साइसर पहुंचनीयता अन्वेषक"]))
```

```python
max_input_length = 128
max_target_length = 128

source_lang = "en"
target_lang = "hi"


def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples["translation"]]
    targets = [ex[target_lang] for ex in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
preprocess_function(raw_datasets["train"][:1])
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

```python
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
batch_size = 8
learning_rate = 2e-5
weight_decay = 0.01
num_train_epochs = 2
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets["test"],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=10)
```

```python
model.save_pretrained("tf_model/")
```

```python
tokenizer.save_pretrained("tokenizer")
```

## Evaluating model using BleU


```python
!pip install nltk transformers datasets
```

```python
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer
from datasets import load_dataset
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import nltk

# Download nltk data for BLEU calculation if you haven't already
nltk.download('punkt')
```

```python
dataset = raw_datasets['test'].select(range(100))
```

```python
smooth_fn = SmoothingFunction().method1

# List to store BLEU scores
bleu_scores = []

# Iterate through each example in the subset
for example in dataset:
    source_text = example['translation']['en']  # Source text (e.g., English)
    reference_text = example['translation']['hi']  # Reference translation (e.g., French)

    # Tokenize and generate the translation
    inputs = tokenizer(source_text, return_tensors="tf", padding=True, truncation=True)
    outputs = model.generate(**inputs)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Tokenize the reference and hypothesis for BLEU calculation
    reference = [nltk.word_tokenize(reference_text)]
    hypothesis = nltk.word_tokenize(generated_text)

    # Calculate BLEU score
    bleu_score = sentence_bleu(reference, hypothesis, smoothing_function=smooth_fn)
    bleu_scores.append(bleu_score)

    print(f"Source: {source_text}")
    print(f"Reference: {reference_text}")
    print(f"Generated: {generated_text}")
    print(f"BLEU Score: {bleu_score}\n")

```

```python
# Calculate the average BLEU score across the subset
average_bleu = sum(bleu_scores) /100
print(f"Average BLEU Score for first 100 examples: {average_bleu}")
```
